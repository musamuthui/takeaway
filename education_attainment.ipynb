```python
import pandas as pd

df = pd.read_csv("Education_attainment.csv")

print(df)
```

## b.

```python
female_totals = df[(df["Sex"] == 'F') & (df["Min degree"] == "bachelor's") & (df["Year"] == 1980)]
male_totals = df[(df["Sex"] == "M") & (df["Min degree"] == "bachelor's") & (df["Year"] == 1980)]

print(f"The percentages for women having earned a Bachelor's Degree is {female_totals["total"].iloc[0]} percent")
print(f"The percentages for men having earned a Bachelor's Degree is {male_totals["total"].iloc[0]} percent")
```

## c.

```python
commonly_awarded_levels = df[((df["Year"] >= 2000) & (df["Year"] <= 2010))]
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
print(commonly_awarded_levels)

top_two_levels = commonly_awarded_levels["Min degree"].value_counts().head(2)

print(top_two_levels)       
```

## Analysis for the two most commonly awarded levels of educational attainment awarded between 2000 and 2010 inclusive
Between 2000 and 2010, the two most commonly awarded education levels were high school and associate’s degrees (21 records each).  
High school completion remained very high throughout the decade (mid-80% to low-90%), with females and white/Asian groups consistently showing the highest rates. Changes over time were small and stable.  
Associate’s degree attainment** was much lower but showed a gradual increase over the decade. Females earned associate’s degrees at higher rates than males, and white/Asian groups tended to have higher percentages than Black and Hispanic groups.  
Overall, high-school completion was stable and near universal, while associate’s degrees showed slow but steady growth with visible demographic differences.



## d.

```python
bachelor_degrees_2000 = df[(df["Year"] == 2000) & (df["Min degree"] == "bachelor's")]
print(bachelor_degrees_2000)

bachelor_degrees_2010 = df[(df["Year"] == 2010) & (df["Min degree"] == "bachelor's")]
print(bachelor_degrees_2010)
```

## How bachelor's degree attainment changed between 2000 and 2010
Between 2000 and 2010, bachelor's degree attainment increased overall. Between 2000 and 2010, bachelor’s degree attainment increased overall. Total attainment rose from 29.1% to 31.7%.   
Women showed the largest growth (30.1% → 35.7%), while men remained about the same. Across racial groups, attainment increased for Whites, Blacks, Hispanics, and American Indians, with Hispanics showing one of the biggest gains.   
Overall, the decade shows steady improvement in bachelor’s degree attainment, driven mainly by women and several minority groups.
